#### Подробное исследование колоннок

```python
def bound(df: pd.DataFrame, column: str) -> pd.DataFrame:
    '''Функция для расчета допустимого интервала, для удаления выбросов на основе межквартильного размаха'''
    df = df.copy()
    series = df[column]
    
    
    q3, q1 = np.percentile (series, [75 ,25])
    iqr = q3 - q1
    lower_bound, upper_bound = q1 - iqr * 1.5, q3 + iqr * 1.5
    share_of_emissions = round(100 - series.loc[
        (lower_bound <= series) & (series <= upper_bound)].shape[0] / series.shape[0] * 100, 2)
    print(f'Верхняя граница = {upper_bound}')
    print(f'Нижняя граница = {lower_bound}')
    print(f'Доля выбросов = {share_of_emissions}%')
    
    if share_of_emissions > 0:
        emission = df.loc[(lower_bound > df[column]) | (df[column] > upper_bound), 'is_canceled']
        emissions_sum, emission_count = emission.sum(), emission.count()

        print(f'{emissions_sum} заказа из {emission_count} аномальных были отменены.')
        df.loc[
            (
                (lower_bound > df[column]) | (df[column] > upper_bound)
            ) & df['is_canceled'] == False
            , column
        ] = df[column].median()
    
    return df
```

## Шаг 3. Формулировка ML-задачи на основе бизнес-задачи

### Введение

Прибыль = Доходы - Расходы, будем производить вычисления на DataFrame с оригинальными данными (df_original)

### Вычисления

####  "Сезонный коэффицент"

В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.

```python
df_original['arrival_date_month'].unique()
```

```python
def seasonal_coefficient(row) -> float:
    spring_autumn = set(['September', 'October', 'November', 'March', 'April', 'May'])
    summer = set(['July', 'August', 'June'])
    
    if row['arrival_date_month'] in spring_autumn:
        return 1.2
    elif row['arrival_date_month'] in summer:
        return 1.4
    else:
        return 1
```

####  "Стоимость номера"

```python
def room_rate_night(row) -> float:
    price = {'A': 1000, 'B': 800, 'C': 600, 'D': 550, 'E': 500, 'F': 450, 'G': 350}
    return seasonal_coefficient(row) * price[row['reserved_room_type']]

def room_rate_service(row) -> float:
    price = {'A': 400, 'B': 350, 'C': 350, 'D': 150, 'E': 150, 'F': 150, 'G': 150}
    return price[row['reserved_room_type']]
```

#### "Обслуживание номера"

```python
def room_service(row) -> float:
    if row['is_canceled']:
        suma = room_rate_service(row) + room_rate_night(row)
    else:
        suma = (1 + row['total_nights'] // 2) * room_rate_service(row)
    return suma
```

#### Добавляем столбец "Прибль"

```python
def profit(row) -> float:
    if row['is_canceled']:
        return - room_service(row)
    else:
        rev =  room_rate_night(row) * row['total_nights'] - room_service(row)
    return rev
```

```python
df_original['profit'] = df_original.apply(profit, axis=1)
```

```python
def ohe(df: pd.DataFrame) -> pd.DataFrame:
    ohe_df = df.copy()

    for column in tqdm(columns):
        one_hot = pd.get_dummies(ohe_df[column])
        ohe_df = ohe_df.drop(columns=[column])
        ohe_df = ohe_df.join(one_hot)
    return ohe_df
```
